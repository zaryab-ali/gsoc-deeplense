```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from tqdm import tqdm

# =========================
# CONFIG
# =========================
TRAIN_DIR = "/content/data/dataset/train"
VAL_DIR   = "/content/data/dataset/val"

BATCH_SIZE = 32
EPOCHS = 30
LR = 1e-3

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# DATASET
# =========================
class LensDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform

        self.class_map = {"no": 0, "sphere": 1, "vort": 2}

        for cls in self.class_map:
            cls_path = os.path.join(root_dir, cls)
            for file in os.listdir(cls_path):
                if file.endswith(".npy"):
                    self.samples.append((os.path.join(cls_path, file), self.class_map[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]

        img = np.load(path)
        img = torch.tensor(img, dtype=torch.float32)

        # Handle all grayscale cases
        if img.ndim == 2:
            img = img.unsqueeze(0)  # [H,W] → [1,H,W]

        if img.shape[0] == 1:
            img = img.repeat(3, 1, 1)  # [1,H,W] → [3,H,W]

        if self.transform:
            img = self.transform(img)

        return img, label

# =========================
# TRANSFORMS
# =========================

mean = 0.061692753184119306
std  = 0.11726841068738307
train_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.Normalize([mean, mean, mean],
                         [std, std, std])
])

val_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.Normalize([mean, mean, mean],
                         [std, std, std])
])

# =========================
# LOAD DATA
# =========================
train_dataset = LensDataset(TRAIN_DIR, transform=train_transform)
val_dataset   = LensDataset(VAL_DIR, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                          shuffle=True, num_workers=2, pin_memory=True)

val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE,
                        shuffle=False, num_workers=2, pin_memory=True)

# =========================
# MODEL
# =========================
model = models.resnet50(pretrained=True)

model.fc = nn.Sequential(
    nn.Linear(model.fc.in_features, 512),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(512, 3)
)

model = model.to(device)

# =========================
# TRAIN SETUP
# =========================
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
scaler = torch.cuda.amp.GradScaler()
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='max',
    factor=0.3,
    patience=5,
    threshold=1e-3,   # ignore tiny fluctuations
    min_lr=1e-6,
    #verbose=True
)

# =========================
# TRAIN FUNCTION
# =========================
def train_one_epoch(epoch):
    model.train()
    total_loss = 0

    print(f"\n===== Epoch {epoch+1}/{EPOCHS} =====")

    for images, labels in tqdm(train_loader):
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f"Train Loss: {avg_loss:.4f}")

    return avg_loss

# =========================
# EVALUATION + ROC
# =========================
def evaluate():
    model.eval()

    all_probs = []
    all_labels = []

    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc="Validating", leave=False):
            images = images.to(device)

            outputs = model(images)
            probs = torch.softmax(outputs, dim=1).cpu().numpy()

            all_probs.append(probs)
            all_labels.append(labels.numpy())

    y_score = np.vstack(all_probs)
    y_true  = np.concatenate(all_labels)

    y_true_bin = label_binarize(y_true, classes=[0,1,2])

    auc = roc_auc_score(y_true_bin, y_score, average="macro")
    print(f"Validation AUC: {auc:.4f}", flush=True)

    return auc, y_true_bin, y_score

def plot_roc(y_true_bin, y_score):
    for i in range(3):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        plt.plot(fpr, tpr, label=f"Class {i}")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.show()
# =========================
# TRAIN LOOP
# =========================
best_auc = 0

for epoch in range(EPOCHS):
    train_one_epoch(epoch)

    val_auc, y_true_bin, y_score = evaluate()
    torch.cuda.empty_cache()

    old_lr = optimizer.param_groups[0]['lr']

    scheduler.step(val_auc)
    new_lr = optimizer.param_groups[0]['lr']
    if new_lr != old_lr:
      print(f"LR reduced: {old_lr:.6f} → {new_lr:.6f}")

    if val_auc > best_auc:
        best_auc = val_auc
        torch.save(model.state_dict(), "best_model.pth")
        print("✅ Saved Best Model")

print(f"\n🔥 Best AUC: {best_auc:.4f}")
model.load_state_dict(torch.load("best_model.pth"))
_, y_true_bin, y_score = evaluate()
plot_roc(y_true_bin, y_score)
```
