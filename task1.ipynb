```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from tqdm import tqdm

# =========================
# CONFIG
# =========================
TRAIN_DIR = "/content/data/dataset/train"
VAL_DIR   = "/content/data/dataset/val"

BATCH_SIZE = 32
EPOCHS = 30
LR = 1e-3

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# DATASET
# =========================
class LensDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform

        self.class_map = {"no": 0, "sphere": 1, "vort": 2}

        for cls in self.class_map:
            cls_path = os.path.join(root_dir, cls)
            for file in os.listdir(cls_path):
                if file.endswith(".npy"):
                    self.samples.append((os.path.join(cls_path, file), self.class_map[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]

        img = np.load(path)
        img = torch.tensor(img, dtype=torch.float32)

        # Handle all grayscale cases
        if img.ndim == 2:
            img = img.unsqueeze(0)  # [H,W] → [1,H,W]

        if img.shape[0] == 1:
            img = img.repeat(3, 1, 1)  # [1,H,W] → [3,H,W]

        if self.transform:
            img = self.transform(img)

        return img, label

# =========================
# TRANSFORMS
# =========================

mean = 0.061692753184119306
std  = 0.11726841068738307
train_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.Normalize([mean, mean, mean],
                         [std, std, std])
])

val_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.Normalize([mean, mean, mean],
                         [std, std, std])
])

# =========================
# LOAD DATA
# =========================
train_dataset = LensDataset(TRAIN_DIR, transform=train_transform)
val_dataset   = LensDataset(VAL_DIR, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                          shuffle=True, num_workers=2, pin_memory=True)

val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE,
                        shuffle=False, num_workers=2, pin_memory=True)

# =========================
# MODEL
# =========================
model = models.resnet50(pretrained=True)

model.fc = nn.Sequential(
    nn.Linear(model.fc.in_features, 512),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(512, 3)
)

model = model.to(device)

# =========================
# TRAIN SETUP
# =========================
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
scaler = torch.cuda.amp.GradScaler()
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='max',
    factor=0.3,
    patience=5,
    threshold=1e-3,   # ignore tiny fluctuations
    min_lr=1e-6,
    #verbose=True
)

# =========================
# TRAIN FUNCTION
# =========================
def train_one_epoch(epoch):
    model.train()
    total_loss = 0

    print(f"\n===== Epoch {epoch+1}/{EPOCHS} =====")

    for images, labels in tqdm(train_loader):
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f"Train Loss: {avg_loss:.4f}")

    return avg_loss

# =========================
# EVALUATION + ROC
# =========================
def evaluate():
    model.eval()

    all_probs = []
    all_labels = []

    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc="Validating", leave=False):
            images = images.to(device)

            outputs = model(images)
            probs = torch.softmax(outputs, dim=1).cpu().numpy()

            all_probs.append(probs)
            all_labels.append(labels.numpy())

    y_score = np.vstack(all_probs)
    y_true  = np.concatenate(all_labels)

    y_true_bin = label_binarize(y_true, classes=[0,1,2])

    auc = roc_auc_score(y_true_bin, y_score, average="macro")
    print(f"Validation AUC: {auc:.4f}", flush=True)

    return auc, y_true_bin, y_score

def plot_roc(y_true_bin, y_score):
    for i in range(3):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        plt.plot(fpr, tpr, label=f"Class {i}")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.show()
# =========================
# TRAIN LOOP
# =========================
best_auc = 0

for epoch in range(EPOCHS):
    train_one_epoch(epoch)

    val_auc, y_true_bin, y_score = evaluate()
    torch.cuda.empty_cache()

    old_lr = optimizer.param_groups[0]['lr']

    scheduler.step(val_auc)
    new_lr = optimizer.param_groups[0]['lr']
    if new_lr != old_lr:
      print(f"LR reduced: {old_lr:.6f} → {new_lr:.6f}")

    if val_auc > best_auc:
        best_auc = val_auc
        torch.save(model.state_dict(), "best_model.pth")
        print("✅ Saved Best Model")

print(f"\n🔥 Best AUC: {best_auc:.4f}")
model.load_state_dict(torch.load("best_model.pth"))
_, y_true_bin, y_score = evaluate()
plot_roc(y_true_bin, y_score)
```

```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models
import torchvision.transforms.v2 as T
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from tqdm import tqdm

# =========================
# CONFIG
# =========================
TRAIN_DIR = "/content/data/dataset/train"
VAL_DIR   = "/content/data/dataset/val"

BATCH_SIZE = 64
EPOCHS = 30
LR = 1e-3
WEIGHT_DECAY = 1e-4
NUM_CLASSES = 3
IMG_SIZE = 150

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# DATASET
# =========================
class LensDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        self.class_map = {"no": 0, "sphere": 1, "vort": 2}

        for cls_name, cls_idx in self.class_map.items():
            cls_path = os.path.join(root_dir, cls_name)
            for file in os.listdir(cls_path):
                if file.endswith(".npy"):
                    self.samples.append((os.path.join(cls_path, file), cls_idx))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = np.load(path).astype(np.float32)

        if img.ndim == 3 and img.shape[0] == 1:
            img = img[0]  # (1, H, W) -> (H, W)

        img = torch.from_numpy(img).unsqueeze(0)  # [1, H, W]

        if self.transform:
            img = self.transform(img)

        return img, label

# =========================
# OPTIONAL: COMPUTE TRAIN MEAN/STD
# =========================
def compute_mean_std(dataset):
    loader = DataLoader(dataset, batch_size=64, shuffle=False, num_workers=2)
    mean = 0.0
    var = 0.0
    n = 0

    for x, _ in loader:
        x = x.float()
        b = x.size(0)
        x = x.view(b, -1)
        mean += x.mean(dim=1).sum().item()
        var += x.var(dim=1, unbiased=False).sum().item()
        n += b

    mean /= n
    std = (var / n) ** 0.5
    return mean, std

# Create a temporary dataset just to compute stats
temp_ds = LensDataset(TRAIN_DIR, transform=None)
train_mean, train_std = compute_mean_std(temp_ds)
print("Train mean:", train_mean, "Train std:", train_std)

# =========================
# TRANSFORMS
# =========================
train_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.RandomHorizontalFlip(p=0.5),
    T.RandomVerticalFlip(p=0.5),
    T.RandomAffine(
        degrees=90,
        translate=(0.05, 0.05),
        scale=(0.9, 1.1),
        shear=None
    ),
    T.RandomRotation(degrees=90),
    T.Normalize(mean=[train_mean], std=[train_std]),
])

val_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.Normalize(mean=[train_mean], std=[train_std]),
])

train_dataset = LensDataset(TRAIN_DIR, transform=train_transform)
val_dataset   = LensDataset(VAL_DIR, transform=val_transform)

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =========================
# MODEL
# =========================
weights = models.ResNet18_Weights.DEFAULT
model = models.resnet18(weights=weights)

# Replace first conv layer to accept 1-channel input
old_conv = model.conv1
model.conv1 = nn.Conv2d(
    1, old_conv.out_channels,
    kernel_size=old_conv.kernel_size,
    stride=old_conv.stride,
    padding=old_conv.padding,
    bias=False
)

# Initialize new conv using averaged pretrained RGB weights
with torch.no_grad():
    model.conv1.weight.copy_(old_conv.weight.mean(dim=1, keepdim=True))

# Replace classifier
model.fc = nn.Sequential(
    nn.Dropout(0.3),
    nn.Linear(model.fc.in_features, NUM_CLASSES)
)

model = model.to(device)

# =========================
# LOSS / OPTIMIZER / SCHEDULER
# =========================
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)

# Better than plateau for this kind of fine-tuning
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
    optimizer,
    T_max=EPOCHS,
    eta_min=1e-6
)

scaler = torch.cuda.amp.GradScaler()

# =========================
# TRAIN / EVAL
# =========================
def train_one_epoch():
    model.train()
    total_loss = 0.0

    for images, labels in tqdm(train_loader, desc="Training", leave=False):
        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    return total_loss / len(train_loader)

@torch.no_grad()
def evaluate():
    model.eval()
    all_probs = []
    all_labels = []

    for images, labels in tqdm(val_loader, desc="Validating", leave=False):
        images = images.to(device, non_blocking=True)
        outputs = model(images)
        probs = torch.softmax(outputs, dim=1).cpu().numpy()

        all_probs.append(probs)
        all_labels.append(labels.numpy())

    y_score = np.vstack(all_probs)
    y_true = np.concatenate(all_labels)
    y_true_bin = label_binarize(y_true, classes=[0, 1, 2])

    auc = roc_auc_score(y_true_bin, y_score, average="macro", multi_class="ovr")
    return auc, y_true_bin, y_score

def plot_roc(y_true_bin, y_score):
    plt.figure(figsize=(8, 6))
    for i, name in enumerate(["no", "sphere", "vort"]):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        plt.plot(fpr, tpr, label=f"{name}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.show()

# =========================
# TRAIN LOOP
# =========================
best_auc = 0.0

# Optional: freeze backbone for a few epochs first
for param in model.layer1.parameters():
    param.requires_grad = False
for param in model.layer2.parameters():
    param.requires_grad = False
for param in model.layer3.parameters():
    param.requires_grad = False
for param in model.layer4.parameters():
    param.requires_grad = False

freeze_epochs = 3

for epoch in range(EPOCHS):
    if epoch == freeze_epochs:
        for param in model.parameters():
            param.requires_grad = True

    train_loss = train_one_epoch()
    val_auc, y_true_bin, y_score = evaluate()
    scheduler.step()

    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {train_loss:.4f} | Val AUC: {val_auc:.4f}")

    if val_auc > best_auc:
        best_auc = val_auc
        torch.save(model.state_dict(), "best_model.pth")
        print("Saved best model")

print(f"Best validation AUC: {best_auc:.4f}")

model.load_state_dict(torch.load("best_model.pth", map_location=device))
final_auc, y_true_bin, y_score = evaluate()
print(f"Final AUC: {final_auc:.4f}")
plot_roc(y_true_bin, y_score)
```
