```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models
import torchvision.transforms.v2 as T
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from tqdm import tqdm

# =========================
# CONFIG
# =========================
TRAIN_DIR = "/content/data/dataset/train"
VAL_DIR   = "/content/data/dataset/val"

BATCH_SIZE = 64
EPOCHS = 20 
LR = 1e-4   
WEIGHT_DECAY = 1e-4
NUM_CLASSES = 3
IMG_SIZE = 150

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ===========
# DATASET
# ====================
class LensDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.samples = []
        self.transform = transform
        self.class_map = {"no": 0, "sphere": 1, "vort": 2}

        for cls_name, cls_idx in self.class_map.items():
            cls_path = os.path.join(root_dir, cls_name)
            if not os.path.exists(cls_path):
                continue
            for file in os.listdir(cls_path):
                if file.endswith(".npy"):
                    self.samples.append((os.path.join(cls_path, file), cls_idx))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = np.load(path).astype(np.float32)

        if img.ndim == 3 and img.shape[0] == 1:
            img = img[0]

        img = torch.from_numpy(img).unsqueeze(0)

        img = img.repeat(3, 1, 1) # [3, H, W]

        if self.transform:
            img = self.transform(img)

        return img, label

# =========================
# TRANSFORMS
# =========================
# FIX: Use standard ImageNet normalization to match the pretrained weights, i have no idea why but this works better than my own data mean and std
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std = [0.229, 0.224, 0.225]

train_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.RandomHorizontalFlip(p=0.5),
    T.RandomVerticalFlip(p=0.5),
    T.RandomAffine(
        degrees=90,
        translate=(0.05, 0.05),
        scale=(0.9, 1.1),
        shear=None
    ),
    T.Normalize(mean=imagenet_mean, std=imagenet_std),
])

val_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.Normalize(mean=imagenet_mean, std=imagenet_std),
])

train_dataset = LensDataset(TRAIN_DIR, transform=train_transform)
val_dataset   = LensDataset(VAL_DIR, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# =========================
# MODEL
# =========================
weights = models.ResNet18_Weights.DEFAULT
model = models.resnet18(weights=weights)

model.fc = nn.Sequential(
    nn.Dropout(0.3),
    nn.Linear(model.fc.in_features, NUM_CLASSES)
)

model = model.to(device)

# =========================
# LOSS / OPTIMIZER / SCHEDULER
# =========================
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    factor=0.1,
    patience=3
    #verbose=True commented because colab was using older version of where this wasnt an argument
)

scaler = torch.cuda.amp.GradScaler()

# =========================
# TRAIN / EVAL
# =========================
def train_one_epoch():
    model.train()
    total_loss = 0.0

    for images, labels in tqdm(train_loader, desc="Training", leave=False):
        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    return total_loss / len(train_loader)

@torch.no_grad()
def evaluate():
    model.eval()
    total_loss = 0.0
    all_probs = []
    all_labels = []

    for images, labels in tqdm(val_loader, desc="Validating", leave=False):
        images = images.to(device, non_blocking=True)
        labels_device = labels.to(device, non_blocking=True)

        outputs = model(images)
        loss = criterion(outputs, labels_device)
        total_loss += loss.item()

        probs = torch.softmax(outputs, dim=1).cpu().numpy()
        all_probs.append(probs)
        all_labels.append(labels.numpy())

    avg_val_loss = total_loss / len(val_loader)

    y_score = np.vstack(all_probs)
    y_true = np.concatenate(all_labels)
    y_true_bin = label_binarize(y_true, classes=[0, 1, 2])

    auc = roc_auc_score(y_true_bin, y_score, average="macro", multi_class="ovr")
    return avg_val_loss, auc, y_true_bin, y_score

def plot_roc(y_true_bin, y_score):
    plt.figure(figsize=(8, 6))
    for i, name in enumerate(["No Substructure", "Spherical", "Vortex"]):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        plt.plot(fpr, tpr, label=f"{name}")
    plt.plot([0, 1], [0, 1], 'k--') # Add diagonal random guessing line
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.show()

# =========================
# TRAIN LOOP
# =========================
best_auc = 0.0

for epoch in range(EPOCHS):
    train_loss = train_one_epoch()
    val_loss, val_auc, y_true_bin, y_score = evaluate()

    scheduler.step(val_loss)

    print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val AUC: {val_auc:.4f}")

    if val_auc > best_auc:
        best_auc = val_auc
        torch.save(model.state_dict(), "best_model.pth")
        print("--> Saved best model")

print(f"\nBest validation AUC: {best_auc:.4f}")

# Evaluate best model
model.load_state_dict(torch.load("best_model.pth", map_location=device))
_, final_auc, y_true_bin, y_score = evaluate()
print(f"Final Model AUC: {final_auc:.4f}")
plot_roc(y_true_bin, y_score)
```
