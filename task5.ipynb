```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms
import torch.nn as nn
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

train_lenses_dir = "/content/data/train_lenses"
train_nonlenses_dir = "/content/data/train_nonlenses"
test_lenses_dir = "/content/data/test_lenses"
test_nonlenses_dir = "/content/data/test_nonlenses"

batch_size = 512     
lr = 1e-3           
num_epochs = 50     
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class LensingDataset(Dataset):
    def __init__(self, lenses_dir, nonlenses_dir, transform=None):
        # List .npy files in each directory
        self.lenses_files = [os.path.join(lenses_dir, f) for f in os.listdir(lenses_dir) if f.endswith(".npy")]
        self.nonlenses_files = [os.path.join(nonlenses_dir, f) for f in os.listdir(nonlenses_dir) if f.endswith(".npy")]
        # Combine file paths and labels: 1 for lens, 0 for non-lens
        self.filepaths = self.lenses_files + self.nonlenses_files
        self.labels = [1]*len(self.lenses_files) + [0]*len(self.nonlenses_files)
        self.transform = transform

    def __len__(self):
        return len(self.filepaths)

    def __getitem__(self, idx):
        # Load the .npy array (shape: 3x64x64)
        arr = np.load(self.filepaths[idx])
        image = torch.from_numpy(arr).float()  
        label = self.labels[idx]
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.float32)



train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(180),    
    transforms.Lambda(lambda x: x + 0.01*torch.randn_like(x)),
    transforms.Lambda(lambda x: torch.clamp(x, 0.0, 1.0))
])

test_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.ToTensor(),
])

num_lenses = len([f for f in os.listdir(train_lenses_dir) if f.endswith(".npy")])
num_nonlenses = len([f for f in os.listdir(train_nonlenses_dir) if f.endswith(".npy")])
print(f"Training: {num_lenses} lenses, {num_nonlenses} non-lenses")

# Create dataset objects
train_dataset = LensingDataset(train_lenses_dir, train_nonlenses_dir, transform=train_transform)
test_dataset  = LensingDataset(test_lenses_dir, test_nonlenses_dir, transform=None)  # no augment on test

# -- WeightedRandomSampler to oversample minority class --
class_counts = np.array([num_nonlenses, num_lenses])
weights_per_class = 1.0 / class_counts
# Assign weight to each sample by its class
sample_weights = [weights_per_class[label] for label in train_dataset.labels]
sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=sampler, num_workers=4, pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)

# -- Model: EfficientNet-B0 with pretrained weights --
model = efficientnet_b0(weights=EfficientNet_B0_Weights.IMAGENET1K_V1).to(device)
# Replace final layer for binary classification (1 output logit)
model.classifier[1] = nn.Linear(model.classifier[1].in_features, 1)
model = model.to(device)

# -- Loss function with pos_weight to handle imbalance --
pos_weight = torch.tensor([num_nonlenses/num_lenses]).to(device) 
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

# Optimizer and LR scheduler
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='max',        
    factor=0.5,        
    patience=5        
    #verbose=True
)
from tqdm import tqdm 

history = {
    "train_loss": [],
    "val_loss": [],
    "auc": []
}

best_auc = 0.0
best_fpr, best_tpr = None, None

for epoch in range(num_epochs):
    print(f"\n===== Epoch {epoch+1}/{num_epochs} =====")

    # ----------
    # TRAINING
    # -----------------
    model.train()
    train_loss = 0.0

    for images, labels in tqdm(train_loader, desc="Training"):
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)[:, 0]
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(train_loader)
    print(f"Train Loss: {train_loss:.4f}")

    # --------------
    # VALIDATION
    # -------------------
    model.eval()
    val_loss = 0.0

    y_true = []
    y_scores = []

    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc="Validation"):
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)[:, 0]
            loss = criterion(outputs, labels)

            val_loss += loss.item()

            probs = torch.sigmoid(outputs)
            y_scores.extend(probs.cpu().numpy())
            y_true.extend(labels.cpu().numpy())

    val_loss /= len(test_loader)

    auc = roc_auc_score(y_true, y_scores)

    fpr, tpr, thresholds = roc_curve(y_true, y_scores)

    print(f"Val Loss: {val_loss:.4f} | AUC: {auc:.4f}")

    # Save history
    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["auc"].append(auc)

    # Save best model
    if auc > best_auc:
        best_auc = auc
        torch.save(model.state_dict(), "best_model.pth")


    print(f"Val Loss: {val_loss:.4f} | AUC: {auc:.4f}")
    scheduler.step(auc)
    current_lr = optimizer.param_groups[0]['lr']
    print(f"Learning Rate: {current_lr:.6f}")


print(f"\nTraining finished. Best validation AUC = {best_auc:.4f}")

plt.figure()
plt.plot(history["train_loss"], label="Train Loss")
plt.plot(history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Loss over epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.savefig("loss_curve.png")

plt.figure()
plt.plot(history["auc"], label="AUC")
plt.legend()
plt.title("AUC over epochs")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.savefig("auc_curve.png")


plt.figure()
plt.plot(best_fpr, best_tpr, label=f"ROC (AUC = {best_auc:.4f})")
plt.plot([0, 1], [0, 1], linestyle="--")  
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
